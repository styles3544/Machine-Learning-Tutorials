We can see that Japan is the leading country with heighest life expectancy in the whole world, followed by Switzerland.

```python
# visualizing relation between year and life expectancy

plt.scatter(df['Year'], df['Life expectancy'])
plt.title("Year vs Life expectancy")
plt.xlabel("Years")
plt.ylabel("Life expectancy")
```

```python
sns.heatmap(df.corr(), annot=True)
```

We can see from the above two charts that life expectancy has increased over the years. 

Same can be concluded with the correlation between year and life expectancy, as they are positively correlated with a value of 0.84.
